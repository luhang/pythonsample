```python
# requestsのget()メソッドを実行

import requests                             # equestsのインポート
rq = requests.get('http://www.yahoo.co.jp') # get()メソッド実行
print(rq.text)                              # 戻り値を表示
```

```python
# ステータスコードを取得

rq = requests.get('http://www.yahoo.co.jp')
print(rq.status_code)
```

```python
# レスポンスのヘッダー情報を取得
print(rq.headers)
```

```python
# 文字コードのエンコード方式を取得
print(rq.encoding)
```
