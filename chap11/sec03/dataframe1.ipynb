```python
# 3列×5行のデータフレームを作成する

import pandas as pd
df = pd.DataFrame(
         {'A': [10, 20, 30, 40, 50],                       # 列Aとその値
          'B': [0.8, 1.6, 2.4, 4.3, 7.6],                  # 列Bとその値
          'C': [-1, -2.6, -3.5, -4.3, -5.1] },             # 列Cとその値
          index = ['row1', 'row2', 'row3', 'row4', 'row5'] # 行名を設定
)
```

```python
# データフレームに行データを追加する

import pandas as pd
df1 = pd.DataFrame(
         {'A': [10, 20, 30, 40, 50],               # 列Aとその値
          'B': [0.8, 1.6, 2.4, 4.3, 7.6],          # 列Bとその値
          'C': [-1, -2.6, -3.5, -4.3, -5.1] },     # 列Cとその値
)
```

```python
df2 = pd.DataFrame(
         {'A': [60, 70, 80, 90, 100],              # 列Aとその値
          'B': [10.2, 11.6, 12.4, 14.3, 17.6],     # 列Bとその値
          'C': [-6, -12.6, -13.5, -14.3, -15.1] }, # 列Cとその値
)
```

```python
df1.append(df2)  # df1にdf2を追加する
```

```python
# 行インデックスが連続するようにして行データを追加する
df2.append(df2, ignore_index=True)  # 行インデックスを連続させる
```

```python
# 行名を設定して追加する

import pandas as pd
df1 = pd.DataFrame(
         {'A': [10, 20, 30, 40, 50],               # 列Aとその値
          'B': [0.8, 1.6, 2.4, 4.3, 7.6],          # 列Bとその値
          'C': [-1, -2.6, -3.5, -4.3, -5.1] },     # 列Cとその値
          index = ['r1', 'r2', 'r3', 'r4', 'r5']   # 行名を設定
)
df2 = pd.DataFrame(
         {'A': [60, 70, 80, 90, 100],              # 列Aとその値
          'B': [10.2, 11.6, 12.4, 14.3, 17.6],     # 列Bとその値
          'C': [-6, -12.6, -13.5, -14.3, -15.1] }, # 列Cとその値
          index = ['r6', 'r7', 'r8', 'r9', 'r10']  # 行名を設定
)
```

```python
df1.append(df2)  # df1にdf2を追加する
```
